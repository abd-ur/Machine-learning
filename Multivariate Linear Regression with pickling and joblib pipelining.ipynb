Multivariate linear regression  with pickiling and joblib pipelining example.

Dataset contains experience, test score and interview score of candidates and shows thier income decided on basis of these features.

```python
import pandas as p
data=p.read_csv('https://raw.githubusercontent.com/Rhevathi/DATA-SCIENCE/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv')
data
```

```python
#Performing data imputation
data['test_score(out of 10)']=data['test_score(out of 10)'].fillna(data['test_score(out of 10)'].mean())
data
```

```python
data=data.dropna()
data
```

```python
#conversion of string to integer 
from word2number import w2n
```

```python
for x in range(2,len(data['experience'])+2):
  data['experience'][x]=w2n.word_to_num(data['experience'][x])
data
```

```python
#linear regression application
from sklearn import linear_model
mod=linear_model.LinearRegression()
mod.fit(data.drop('salary($)',axis='columns'),data['salary($)'])
print(mod.predict([[2,9,6]]))
print(mod.predict([[12,10,10]]))
```

```python
#pickle
import pickle
with open('mod_pkl','wb') as f:
  pickle.dump(mod,f)
```

```python
with open('mod_pkl','rb') as d:
  srg=pickle.load(d)
srg.predict([[1,10,11]])
```

```python
#joblib
import joblib as jb
jb.dump(mod,'mod_jb')
jb_ob=jb.load('mod_jb')
```

```python
jb_ob.predict([[3,4,5]])
```
